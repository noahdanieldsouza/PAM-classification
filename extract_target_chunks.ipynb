```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# === Parameters ===
input_path ='/content/drive/MyDrive/mass_data/8567.240730165948.wav' # your 3-hour file
output_dir ='/content/drive/MyDrive/mass_data/cut2'         # folder to save chunks
chunk_duration = 5  # seconds

# Create output folder if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# === Load the entire audio file ===
y, sr = librosa.load(input_path, sr=None)
total_duration = librosa.get_duration(y=y, sr=sr)
samples_per_chunk = int(chunk_duration * sr)

print(f"Loaded file: {input_path}")
print(f"Total duration: {total_duration / 60:.2f} minutes")
print(f"Saving 5-second chunks to: {output_dir}")

# === Split and save ===
for i in range(0, len(y), samples_per_chunk):
    chunk = y[i:i + samples_per_chunk]

    start_sec = i / sr
    end_sec = min((i + samples_per_chunk) / sr, total_duration)

    filename = f"2.10chunk_{int(start_sec):06d}_{int(end_sec):06d}.wav"
    out_path = os.path.join(output_dir, filename)

    sf.write(out_path, chunk, sr)
    print(f"Saved: {filename}")

print("\nâœ… All chunks saved.")
```
